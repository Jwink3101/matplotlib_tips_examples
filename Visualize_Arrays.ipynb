## Cleaned Up

```python
fig,ax = plt.subplots(figsize=(4,7),dpi=100)

pp = ax.imshow(data,vmin=0,vmax=1,cmap=plt.cm.Greys)

ylabels = ['ylabel (%i)' %i for i in range(20)]
xlabels = ['xlabel (%i)' %i for i in range(10)]

Ny = len(ylabels)
Nx = len(xlabels)

ax.set_xticks(range(Nx))
ax.set_xticklabels(xlabels,rotation=90,fontsize=10)

ax.set_yticks(range(Ny))
ax.set_yticklabels(ylabels,rotation=0,fontsize=10)

# use minor ticks for grid lines https://stackoverflow.com/a/38994970
ax.set_xticks(np.arange(-.5, Nx, 1), minor=True);
ax.set_yticks(np.arange(-.5, Ny, 1), minor=True);

# # Gridlines based on minor ticks
ax.grid(which='minor', color=[0.5]*3, linestyle='-', linewidth=1) # If you do something like reds, maybe do 0.1

from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
cbar = fig.colorbar(pp,cax=cax)


ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
