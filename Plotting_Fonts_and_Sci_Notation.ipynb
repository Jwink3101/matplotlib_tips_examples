# Plotting Fonts & Scientific Notation

Set different plot styles

````python
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline

import sys

print('Matplotlib Version: ' + mpl.__version__)
print('NumPy Version: ' + np.version.full_version)
print('Python Version: ' + sys.version)

import datetime
now  = datetime.datetime.now().isoformat()
print('Ran on ' + now)
````

````python
x = np.linspace(0,2*np.pi,10**3)
y = np.sin(x) + np.cos(x)
````

## Basics

````python
fig,ax = plt.subplots()
ax.plot(x,y,'-k')
ax.set_xlabel(r'x label with math (ex: $x^2$)')
ax.set_ylabel(r'y label with mathregular(ex: $\mathregular{{x^2}}$)'); # Doesn't work for LaTeX render
````

### Default *Specified*

````python
plt.rc('font', family='sans-serif') # These are the defaults
plt.rc('text', usetex=False)

fig,ax = plt.subplots()
ax.plot(x,y,'-k')
ax.set_xlabel(r'x label with math (ex: $x^2$)')
ax.set_ylabel(r'y label with mathregular(ex: $\mathregular{{x^2}}$)'); # Doesn't work for LaTeX render
````

## Serif

````python
plt.rc('font', family='serif')

fig,ax = plt.subplots()
ax.plot(x,y,'-k')
ax.set_xlabel(r'x label with math (ex: $x^2$)')
ax.set_ylabel(r'y label with mathregular(ex: $\mathregular{{x^2}}$)'); # Doesn't work for LaTeX render
````

## LaTeX

LaTeX must be installed

````python
plt.rc('font', family='serif')
plt.rc('text', usetex=True)

fig,ax = plt.subplots()
ax.plot(x,y,'-k')
ax.set_xlabel(r'x label with math (ex: $x^2$)')
ax.set_ylabel('Mathregular is not supported')
#ax.set_ylabel(r'y label with mathregular(ex: $\mathregular{{x^2}}$)'); # Doesn't work for LaTeX render
````

````python
# Reset it before the next section
plt.rc('font', family='sans-serif')
plt.rc('text', usetex=False)
````

## Scientific Notation

See other notebooks where this is also set as defaults. Specifically

```
axes.formatter.limits : -3, 3       # When to use scientific notation
axes.formatter.use_mathtext : True  # False:1e6 vs True: 1 \times 10^6
```

````python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(7,3),dpi=100,num=1)

# Regular
ax1.plot(x,1e4*y)

# Scientific
ax2.plot(x,1e4*y)

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1))

ax2.yaxis.set_major_formatter(formatter)

fig.tight_layout()
````
